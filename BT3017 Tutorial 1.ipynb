### d)

- Match of England vs Germany

```python
countries2 = ["England", "Germany"]

df6 = df[df["home_team"].isin(countries2) & df["away_team"].isin(countries2)]
print("Length of DataFrame df6 is",len(df6.index))
df6
```

```python
df6.isnull().sum() # Check for any incomplete data before analysis
```

#### i) 

- Do a scatterplot of the data contained in df6. 
- x - home_score; y - away_score
- hue - home_team

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.scatterplot(x = "home_score", y = "away_score", data=df6, hue = "home_team", ax=ax)
plt.legend(loc='upper right')
```

#### ii)

- Maximum number of goals scored by one team in all the England vs Germany matches

```python
# Get the max of home_score or away_score

max(df6["home_score"].max(), df6["away_score"].max())
```

#### iii)

- Print a heatmap of home vs away scores for England vs Germany matches
- Heatmap axes follow the same convention as `part i)` x: home_score; y: away_score

```python
# Long format

df7 = df6[["home_score","away_score"]].groupby(["home_score","away_score"]).size()
df7
```

```python
# Long to wide format

df7 = df7.unstack().fillna(0)
df7.insert(4, 4, pd.DataFrame([0 for i in range(len(df7))]))
df7 = df7.transpose() # Using the same convention as part (i) x: home_score; y: away_score
df7
```

```python
sns.heatmap(df7, annot=True)

# Ensure the x-axis & y-axis start from 0
plt.ylim(0, df6["away_score"].max() + 1)
plt.xlim(0, df6["home_score"].max() + 1)
```

#### iv) 

- Compute England’s 
    - home win rate <sup>2</sup>, 
    - away win rate <sup>2</sup>, 
    - and total win rate <sup>2</sup> against Germany.
    
<sup>2</sup> win rate is strictly defined as **Wins divided by the total number of matches played (Wins, Draws, Losses)**

```python
# Home win rate
# Home wins / Total Home for England

dfEngH = df6[df6["home_team"] == "England"]
len(dfEngH[dfEngH["home_score"] > dfEngH["away_score"]]) / len(dfEngH)
```

```python
# Away win rate
# Away wins / Total Away for England

dfEngA = df6[df6["away_team"] == "England"]
len(dfEngA[dfEngA["away_score"] > dfEngA["home_score"]]) / len(dfEngA)
```

```python
# Total win rate for England
# (Wins Home + Wins Away) / Total England vs Germany i.e. len(df6)

EngWins = len(dfEngH[dfEngH["home_score"] > dfEngH["away_score"]]) + len(dfEngA[dfEngA["away_score"] > dfEngA["home_score"]])

EngWins / len(df6)
```
